```python
from PIL import Image

import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
image_ttls = os.listdir('data_yael')
```

```python
list_of_images = [np.asarray(Image.open(f'data_yael/{img_title}')) for img_title in image_ttls]
```

```python
len(list_of_images) #num of images
```

```python
plt.figure(figsize=(10,8))
for i in range(15):
    plt.subplot(3,5,i+1)
    plt.imshow(list_of_images[i])
    plt.xticks([], [])
    plt.yticks([], [])
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
for i in range(15, 30):
    plt.subplot(3,5,i-15+1)
    plt.imshow(list_of_images[i])
    plt.xticks([], [])
    plt.yticks([], [])
plt.tight_layout()
```

```python
d,  d_n = np.unique([im_array.ndim for im_array in list_of_images], return_counts=True)
dict(zip(list(d), list(d_n)))# Counts of width of images 
```

### 90 images are black and white, 10 images are color images

```python
h, h_n = np.unique([im_array.shape[0] for im_array in list_of_images], return_counts=True)
dict(zip(list(h), list(h_n))) # Counts of height of images 
```

```python
w, w_n = np.unique([im_array.shape[1] for im_array in list_of_images], return_counts=True)
dict(zip(list(w), list(w_n))) # Counts of width of images 
```

```python
p,  p_n = np.unique([im_array.shape[0]/im_array.shape[1] for im_array in list_of_images], return_counts=True)
dict(zip(list(p), list(p_n))) # Counts of proportion of width to height
```

## Images are of different size. 90 of them are square 28*28 and the rest is rectangular

```python
mn,  mn_n = np.unique([im_array.min() for im_array in list_of_images], return_counts=True)
dict(zip(list(mn), list(mn_n)))
```

```python
mx,  mx_n = np.unique([im_array.max() for im_array in list_of_images], return_counts=True)
dict(zip(list(mx), list(mx_n)))
```
